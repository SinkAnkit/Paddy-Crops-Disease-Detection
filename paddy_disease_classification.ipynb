```python
import os
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.applications import VGG16
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
from google.colab import drive
import gc

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Mount Google Drive
drive.mount('/content/drive')

# Paths
BASE_DIR = "/content/paddy-disease-classification"
DRIVE_SAVE_DIR = "/content/drive/MyDrive/paddy_models/"
TRAIN_DIR = os.path.join(BASE_DIR, "train_images")
TRAIN_CSV = os.path.join(BASE_DIR, "train.csv")

# Create save directory if it doesn't exist
if not os.path.exists(DRIVE_SAVE_DIR):
    os.makedirs(DRIVE_SAVE_DIR)

# Image parameters
IMG_SIZE = (224, 224)
NUM_CLASSES = 10
BATCH_SIZE = 32
EPOCHS = 30

# Check if GPU is available
print("GPU Available:", tf.config.list_physical_devices('GPU') != [])

# Function to prepare data
def prepare_data():
    """Load and split dataset into training and validation sets, create data generators."""
    df = pd.read_csv(TRAIN_CSV)
    df['image_path'] = df.apply(lambda row: os.path.join(TRAIN_DIR, row['label'], row['image_id']), axis=1)

    # Split into train and validation
    train_df = df.sample(frac=0.8, random_state=42)
    val_df = df.drop(train_df.index)

    # Data augmentation and preprocessing
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True
    )
    val_datagen = ImageDataGenerator(rescale=1./255)

    train_generator = train_datagen.flow_from_dataframe(
        train_df,
        x_col='image_path',
        y_col='label',
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical'
    )
    val_generator = val_datagen.flow_from_dataframe(
        val_df,
        x_col='image_path',
        y_col='label',
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical'
    )

    return train_generator, val_generator, train_df, val_df

# Function to build Custom CNN
def build_custom_cnn():
    """Build and compile a custom CNN model."""
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(NUM_CLASSES, activation='softmax')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Function to build VGG16
def build_vgg16():
    """Build and compile a VGG16 model with transfer learning."""
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    for layer in base_model.layers[:-4]:  # Unfreeze last 4 layers
        layer.trainable = False

    x = Flatten()(base_model.output)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(NUM_CLASSES, activation='softmax')(x)

    model = Model(inputs=base_model.input, outputs=x)
    model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Function to build HSV CNN
def build_hsv_cnn():
    """Build and compile an HSV CNN model with HSV preprocessing function."""
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(NUM_CLASSES, activation='softmax')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

    def hsv_preprocess(x):
        """Convert image from RGB to HSV."""
        x = x * 255  # Denormalize
        x = cv2.cvtColor(x.astype(np.uint8), cv2.COLOR_RGB2HSV)
        return x / 255.0  # Normalize back

    return model, hsv_preprocess

# Function to train and evaluate models
def train_and_evaluate(model, train_generator, val_generator, train_df, val_df, name, preprocess=None):
    """Train model, plot individual graphs, save model, and return history."""
    if preprocess:
        train_datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            horizontal_flip=True,
            preprocessing_function=preprocess
        )
        val_datagen = ImageDataGenerator(rescale=1./255, preprocessing_function=preprocess)

        train_generator = train_datagen.flow_from_dataframe(
            train_df,
            x_col='image_path',
            y_col='label',
            target_size=IMG_SIZE,
            batch_size=BATCH_SIZE,
            class_mode='categorical'
        )
        val_generator = val_datagen.flow_from_dataframe(
            val_df,
            x_col='image_path',
            y_col='label',
            target_size=IMG_SIZE,
            batch_size=BATCH_SIZE,
            class_mode='categorical'
        )

    history = model.fit(
        train_generator,
        validation_data=val_generator,
        epochs=EPOCHS,
        verbose=1
    )

    # Save the model
    model_save_path = os.path.join(DRIVE_SAVE_DIR, f"{name.replace(' ', '_').lower()}_model.h5")
    model.save(model_save_path)
    print(f"Saved {name} model to {model_save_path}")

    # Plot individual accuracy and loss
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.title(f'{name} Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{name} Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    plt.close()

    # Clear memory
    tf.keras.backend.clear_session()
    gc.collect()

    return history.history

# Main execution
if __name__ == "__main__":
    # Prepare data
    train_generator, val_generator, train_df, val_df = prepare_data()
    print(f"Training samples: {train_generator.samples}, Validation samples: {val_generator.samples}")

    # Train Custom CNN
    print("Starting training for Custom CNN")
    custom_cnn_model = build_custom_cnn()
    custom_history = train_and_evaluate(custom_cnn_model, train_generator, val_generator, train_df, val_df, "Custom CNN")

    # Train VGG16
    print("Starting training for VGG16")
    vgg16_model = build_vgg16()
    vgg_history = train_and_evaluate(vgg16_model, train_generator, val_generator, train_df, val_df, "VGG16")

    # Train HSV CNN
    print("Starting training for HSV CNN")
    hsv_model, preprocess_hsv = build_hsv_cnn()
    hsv_history = train_and_evaluate(hsv_model, train_generator, val_generator, train_df, val_df, "HSV CNN", preprocess_hsv)

    # Collect final metrics
    custom_acc = custom_history['val_accuracy'][-1]
    custom_loss = custom_history['val_loss'][-1]
    vgg_acc = vgg_history['val_accuracy'][-1]
    vgg_loss = vgg_history['val_loss'][-1]
    hsv_acc = hsv_history['val_accuracy'][-1]
    hsv_loss = hsv_history['val_loss'][-1]

    # Print model comparison
    print("\nModel Comparison:")
    print(f"Custom CNN - Val Accuracy: {custom_acc:.4f}, Val Loss: {custom_loss:.4f}")
    print(f"VGG16 - Val Accuracy: {vgg_acc:.4f}, Val Loss: {vgg_loss:.4f}")
    print(f"HSV CNN - Val Accuracy: {hsv_acc:.4f}, Val Loss: {hsv_loss:.4f}")

    # Combined accuracy graph
    plt.figure(figsize=(10, 6))
    plt.plot(custom_history['accuracy'], 'b-', label='Custom CNN Train Acc')
    plt.plot(custom_history['val_accuracy'], 'b--', label='Custom CNN Val Acc')
    plt.plot(vgg_history['accuracy'], 'g-', label='VGG16 Train Acc')
    plt.plot(vgg_history['val_accuracy'], 'g--', label='VGG16 Val Acc')
    plt.plot(hsv_history['accuracy'], 'r-', label='HSV CNN Train Acc')
    plt.plot(hsv_history['val_accuracy'], 'r--', label='HSV CNN Val Acc')
    plt.title('Training and Validation Accuracy Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
    plt.close()

    # Combined loss graph
    plt.figure(figsize=(10, 6))
    plt.plot(custom_history['loss'], 'b-', label='Custom CNN Train Loss')
    plt.plot(custom_history['val_loss'], 'b--', label='Custom CNN Val Loss')
    plt.plot(vgg_history['loss'], 'g-', label='VGG16 Train Loss')
    plt.plot(vgg_history['val_loss'], 'g--', label='VGG16 Val Loss')
    plt.plot(hsv_history['loss'], 'r-', label='HSV CNN Train Loss')
    plt.plot(hsv_history['val_loss'], 'r--', label='HSV CNN Val Loss')
    plt.title('Training and Validation Loss Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    plt.close()

    # Bar graph for final validation accuracy
    models = ['Custom CNN', 'VGG16', 'HSV CNN']
    val_accs = [custom_acc, vgg_acc, hsv_acc]
    plt.figure(figsize=(8, 5))
    bars = plt.bar(models, val_accs, color=['blue', 'green', 'orange'])
    plt.title('Final Validation Accuracy Comparison')
    plt.ylabel('Accuracy')
    for bar, acc in zip(bars, val_accs):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, f'{acc:.4f}', ha='center')
    plt.show()
    plt.close()

    # Bar graph for final validation loss
    val_losses = [custom_loss, vgg_loss, hsv_loss]
    plt.figure(figsize=(8, 5))
    bars = plt.bar(models, val_losses, color=['blue', 'green', 'orange'])
    plt.title('Final Validation Loss Comparison')
    plt.ylabel('Loss')
    for bar, loss in zip(bars, val_losses):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, f'{loss:.4f}', ha='center')
    plt.show()
    plt.close()
```

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
from google.colab import drive
import random
import tensorflow as tf
import gc

# Mount Google Drive
drive.mount('/content/drive')

# Paths
BASE_DIR = "/content/paddy-disease-classification"
DRIVE_SAVE_DIR = "/content/drive/MyDrive/paddy_models/"
TRAIN_DIR = os.path.join(BASE_DIR, "train_images")

# Image parameters
IMG_SIZE = (224, 224)

# Model paths
custom_cnn_path = os.path.join(DRIVE_SAVE_DIR, "custom_cnn_model.h5")
vgg16_path = os.path.join(DRIVE_SAVE_DIR, "vgg16_model.h5")
hsv_cnn_path = os.path.join(DRIVE_SAVE_DIR, "hsv_cnn_model.h5")

# Class labels (based on your dataset with 10 classes)
class_labels = ['bacterial_leaf_blight', 'bacterial_leaf_streak', 'bacterial_panicle_blight',
                'blast', 'brown_spot', 'dead_heart', 'downy_mildew', 'hispa', 'normal', 'tungro']

# Function to load a random image from train_images
def load_random_train_image(train_dir):
    """Load a random image from the training directory and extract its true label."""
    # Get all class subdirectories
    class_dirs = [d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))]
    if not class_dirs:
        raise ValueError(f"No class directories found in {train_dir}")

    # Pick a random class
    random_class = random.choice(class_dirs)
    class_path = os.path.join(train_dir, random_class)

    # Get all images in the class directory
    image_files = [f for f in os.listdir(class_path) if f.endswith(('.jpg', '.jpeg', '.png'))]
    if not image_files:
        raise ValueError(f"No images found in {class_path}")

    # Pick a random image
    random_image = random.choice(image_files)
    image_path = os.path.join(class_path, random_image)

    # Load and preprocess the image
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Failed to load image: {image_path}")
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

    # True label is the class directory name
    true_label = random_class

    return image, image_path, true_label

# Function to preprocess image for prediction
def preprocess_image(image, model_type):
    """Preprocess the image based on the model type (RGB or HSV)."""
    img = cv2.resize(image, IMG_SIZE)
    img = img_to_array(img)

    if model_type == "hsv_cnn":
        # HSV preprocessing
        img = img * 255  # Denormalize
        img = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_RGB2HSV)
        img = img / 255.0  # Normalize back
    else:
        # RGB preprocessing for Custom CNN and VGG16
        img = img / 255.0  # Normalize

    img = np.expand_dims(img, axis=0)  # Add batch dimension
    return img

# Function to predict with a model
def predict_with_model(model_path, image, model_name):
    """Load a model and predict the class of an image."""
    model = load_model(model_path)
    prediction = model.predict(image, verbose=0)
    predicted_class_idx = np.argmax(prediction[0])
    confidence = prediction[0][predicted_class_idx]

    # Clear memory
    tf.keras.backend.clear_session()
    gc.collect()

    return class_labels[predicted_class_idx], confidence

# Function to display image with predictions
def display_results(image, true_label, predictions):
    """Display the image with true label and model predictions above it."""
    plt.figure(figsize=(8, 6))
    plt.imshow(image)
    plt.axis('off')

    # Create title with true label and predictions
    title = f"True Label: {true_label}\n"
    for model_name, (pred_label, confidence) in predictions.items():
        title += f"{model_name}: {pred_label} ({confidence:.4f})\n"

    plt.title(title, fontsize=12, pad=20)
    plt.show()
    plt.close()

# Main execution
if __name__ == "__main__":
    # Load random image from training data
    image, image_path, true_label = load_random_train_image(TRAIN_DIR)
    print(f"Selected image: {image_path}")

    # Preprocess image for each model
    rgb_image = preprocess_image(image, "custom_cnn")  # Same for Custom CNN and VGG16
    hsv_image = preprocess_image(image, "hsv_cnn")

    # Load and predict with each model
    predictions = {}
    predictions["Custom CNN"] = predict_with_model(custom_cnn_path, rgb_image, "Custom CNN")
    predictions["VGG16"] = predict_with_model(vgg16_path, rgb_image, "VGG16")
    predictions["HSV CNN"] = predict_with_model(hsv_cnn_path, hsv_image, "HSV CNN")

    # Display the image with predictions
    display_results(image, true_label, predictions)
```
